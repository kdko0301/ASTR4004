```python
# pkgs

import numpy as np
import matplotlib.pyplot as plt
import astropy.constants as c
import astropy.units as u
```

```python
MW_DM_Mass = 1.5e12 *c.M_sun.value
MW_DM_rad = 200 * c.kpc.value# kpc
Avg_DM_Mass = 2.6e12 *c.M_sun.value 
Avg_DM_rad = 360 * c.kpc.value #kpc
GeV_kg = 1.783e-27 #GeV-kg conversion
MW_den = MW_DM_Mass/(4/3*np.pi*MW_DM_rad**3)
Avg_den = Avg_DM_Mass/(4/3*np.pi*Avg_DM_rad**3)
# WIMP mass range
ms_list = np.logspace(1, 6, 10)
```

```python
def cross_section(fp, nd):
    # fp: mean free path, nd: number density
    sig = 1/nd/fp
    return sig
```

```python
#MW only case
cs_list = cross_section(MW_DM_rad, MW_den/ms_list/GeV_kg)
plt.plot(ms_list, cs_list*1e4)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('WIMP Mass (GeV)')
plt.ylabel('Cross section (cm^2)')
cs_list
```

```python
#MW + source case
cs_list = 1/(1/cross_section(MW_DM_rad, MW_den/ms_list/GeV_kg) + 1/cross_section(Avg_DM_rad, Avg_den/ms_list/GeV_kg))
plt.plot(ms_list, cs_list*1e4)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('WIMP Mass (GeV)')
plt.ylabel('Cross section (cm^2)')

cs_list
```
