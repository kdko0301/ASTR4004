```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
import astropy.constants as c
```

```python
# lorentz transformation
def lorentz(x, y, z, t, v):
    B = v/c
    g = (1-B**2)**(-1/2) #Gamma value
    xp = g*(x-v*t)
    yp = y
    zp = z
    tp = g*(t-v/c**2*x)
    return xp, yp, zp, tp
```

Q1-b

```python
#variables
c = 3e5 #lightspeed in km/s
v = 0.866*c
B = v/c #Beta value (v/c)
g = (1-B**2)**(-1/2) #Gamma value
xlist = [300, 259.8, 150 , 0, -150, -259.8, -300, -259.8, -150, -0, 150, 259.8]
ylist = [0, 150, 259.8, 300, 259.8, 150, 0, -150, -259.8, -300, -259.8, -150]
t_r = 0.001 #reflection time
```

```python
xplist = []
for x in xlist:
    xp, yp, zp, tp = lorentz(x, 0, 0, t_r, v)
    xplist.append(xp)
xplist
```

```python
plt.figure(figsize=(10,4))
plt.plot(xlist, ylist, 'b.', label = 'rest frame corner cube')
plt.plot(xplist, ylist, 'r.', label = 'moving frame corner cube at time of reflection')
plt.legend(bbox_to_anchor=(1, 1))
plt.xlabel('x (km)')
plt.ylabel('y (km)')
```
